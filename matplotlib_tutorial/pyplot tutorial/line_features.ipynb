```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def curves(a, b, c, d):
    x = np.linspace(0.1, 2.0, 20)
    return a*np.power(x, 2) + b*np.log(x) + c*np.sin(x) + d
```

```python
figure1 = plt.figure(1,figsize=(12,12))
sp1 = plt.subplot(3,3,1)
y1 = curves(1,2,3,2)
line1, = plt.plot(y1)
#plt.setp(line1)

sp2 = plt.subplot(3,3,2)
line2, = plt.plot(y1)
plt.setp(line2,animated=False, aa=True)
print('1',line2.axes == sp2)

sp3 = plt.subplot(333)
line3, = plt.plot(y1)
plt.setp(line3,c='r',dash_capstyle='projecting' )
plt.setp(line3, dash_joinstyle='miter', dashes=[1,6,2,10])

sp4 = plt.subplot(334)
line4, =plt.plot(y1, 'c')
plt.setp(line4, drawstyle='steps')

sp5 = plt.subplot(335)
line5, = plt.plot(y1, 'm')
print('2',line5.figure == figure1)
plt.setp(line5, fillstyle='left')
print('3',line5.get_gid())
line5.set_gid('Artist name')
print('4',line5.get_gid())

sp6 = plt.subplot(336)
line6, = plt.plot(y1, 'k.')
plt.setp(line6, fillstyle='left')
plt.setp(line6, label='dot line')
plt.legend()
print('5',line6.get_label())

sp7 = plt.subplot(337)
line7,line7c= plt.plot(y1, 'g',12-y1, 'y')
plt.setp(line7, ls='solid')
plt.setp([line7,line7c], linewidth='4')

sp8 = plt.subplot(338)
sp8.set_axis_bgcolor('k')
line8, =plt.plot(np.sin(y1), 'w')
line8.set_antialiased(True)
line8.set_linestyle('--')
plt.setp(line8, fillstyle='full')
line8.set_marker('+')
line8.set_label('+')
plt.legend()

sp9 = plt.subplot(339)
line9, =plt.plot(-y1,'m')
line9.set_marker('o')
line9.set_markeredgecolor('k')
line9.set_markerfacecolor('b')
line9.set_markeredgewidth(2)
print('6',line9.get_xdata())
line9.set_visible(True)

for x in figure1.axes:
    plt.xlabel('x')
    plt.ylabel('y')
```

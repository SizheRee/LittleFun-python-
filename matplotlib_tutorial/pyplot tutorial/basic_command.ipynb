```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
figure1 = plt.figure(1, figsize=(12, 12), dpi=100, facecolor='c', edgecolor='k')  #figsize (width,height) inches
sp1 = plt.subplot(331)
line1 = plt.plot([1,2,3,4], label = 'line1')
sp2 = plt.subplot(332)
plt.plot([1,2,3,4],[4,3,2,1],'r--')
sp3 = plt.subplot(333)
plt.plot([2.5,4,5,6],[1,3,4,5],'k.')
sp4 = plt.subplot(334)
x = np.linspace(1,5,50)
y = np.sin(x)
plt.plot(x, y, 'b2')
plt.ylabel(r'sin(x)')
plt.xlabel('x') 
sp5 = plt.subplot(335)
plt.plot(x, np.cos(x), 'ms')
plt.ylabel(r'cos(x)')
sp6 = plt.subplot(3,3,6)
plt.plot(x, np.exp(x), 'w*')
sp7 = plt.subplot(337)
plt.plot(x, np.sin(x), 'md', x, np.cos(x), 'k+', x, np.log(x), 'rx')
plt.axis([0,6,-2,2])
sp8 = plt.subplot(338)
plt.subplot(331)
line2 = plt.plot([4,3,2,1], label = 'line2')
plt.legend()
plt.subplot(339)
aline = plt.plot(x, label = 'a line')
plt.ylabel('y')
plt.xlabel('x')
plt.setp(aline, color = 'k', linewidth = 3.0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
```

```python
def f(t):
    return np.exp(t)*np.cos(t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t = np.linspace(0, 5 , 100)
figure1 = plt.figure(1, figsize=(8,4))
sp1 = plt.subplot(121)
sp1.plot([x for x in range(1,50)])
line1,line2 = plt.plot(t, f(t), 'ko',t, 0.3*f(t), 'm-')
sp2 = plt.subplot(122)
line3, = plt.plot(t1, np.cos(2*np.pi*t1), 'r--')

figure2 = plt.figure(2, figsize=(8,4))
sp3 = plt.subplot(121)
plt.plot(t2, np.power(t2,2), 'c-')

sp4 = plt.Subplot(figure2, 122)
sp4.set_visible = True
sp4.plot([1, 2, 3, 4])
sp4.set_xticks([])
sp4.set_yticks([])

figure2.add_subplot(sp4)
sp4.set_frame_on(False)

```

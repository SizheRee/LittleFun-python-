```python
ax = gridspec.GridSpec(3, 3, wspace=0.0, hspace=0.0)
Figure2 = plt.figure(2, figsize=(12, 8))
ax_x, ax_y = ax.get_geometry()

x = np.linspace(0, 5)
for i,j in enumerate((product(range(ax_x),range(ax_y)))):
    sp = plt.Subplot(Figure2, ax[i])
    sp.set_xticks([])
    sp.set_yticks([])
    sp.plot(x, np.cos(x)*np.sin(i*x))
    Figure2.add_axes(sp)
    
gs = gridspec.GridSpec(Figure2.axes[0], (3,3))
```

```python
Figure2 = plt.figure(2, figsize=(12, 8))

def funny(a, b, x = np.linspace(0, 5)):
    return np.sin(np.exp(a*x)) + np.cos(b*x)
outter_grid = gridspec.GridSpec(3, 3, wspace=0.0, hspace=0.0)
out_x, out_y = outter_grid.get_geometry()

img = plt.imread('./pngs/stinkbug.png')
for a, i in enumerate(product(range(out_x),range(out_y))):
    inner_grid = gridspec.GridSpecFromSubplotSpec(3, 3,
                    subplot_spec=outter_grid[i], wspace=0.0, hspace=0.0)
    in_x, in_y = inner_grid.get_geometry()
    for b,j in enumerate(product(range(out_x),range(out_y))):
        sp2 = plt.Subplot(Figure2, inner_grid[j])
        sp2.set_xticks([])
        sp2.set_yticks([])
        sp2.plot([1,2,3,4])
        Figure2.add_axes(sp2)
```

```python
Figure3 = plt.figure(3, figsize=(8,8))
gs = gridspec.GridSpec(8, 8, wspace=0.0, hspace=0.0,
gs_x, gs_y = gs.get_geometry()
img = plt.imread('./pngs/stinkbug.png')
for i in range(gs_x * gs_y):
    ax = plt.Subplot(Figure3, gs[i])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(img)
    Figure3.add_axes(ax)
```

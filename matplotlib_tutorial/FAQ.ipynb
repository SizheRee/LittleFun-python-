```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
a = 1
t = np.arange(-10, 10, 0.1)
x = a*(t - np.sin(t))
y = a*(t - np.cos(t))

plt.plot(x, y, 'b-')
plt.text(-14, 5, 'x=a*(t -sint)')
plt.text(-14, 0, 'y=a*(t-cost)')
```

```python
theta=np.arange(0, 2*np.pi, 0.02)  

Fig1 = plt.figure(1)
gs = gridspec.GridSpec(3, 3, wspace=0, hspace=0)
a = 1
for i in gs:
    plt.subplot(i, polar = True)
    plt.xticks([])
    plt.yticks([])
    plt.plot(theta, a*(1 + np.cos(a*theta)), lw=2, label='a = %s'%a)
    leg = plt.legend(loc = 2, fontsize = 7, markerscale = 0.5, markerfirst = False,
                    shadow = True)
    a = a+1
    
Fig1.text(.4, 1, 'r = a*(1 + cos(a*))')
```

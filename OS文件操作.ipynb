```python
import os
```

```python
path = "I:\电视剧\美剧\飞出个未来 全七季\\"
```

批量重命名

```python
for dirs in os.listdir():
    npath = path + dirs
    print(dirs)
    os.chdir(dirs)
    for i in os.listdir(npath):
        filename, ext = os.path.splitext(i)
        if(ext == '.mp'):
            ext = '.mp4'
            print(filename + ext)
            os.rename(filename+'.mp', filename+ext)
    os.chdir(path)
```

```python
with open("./time.txt", "r") as fp:
    timedata = fp.read()
    
with open("./user.txt", "r", encoding='utf-8') as fp:
    userdata = fp.read()
```

```python
import io
```

```python
tio = io.StringIO(timedata)
uio = io.StringIO(userdata)
```

```python
import re
direname = re.compile("^lab2_\w+")
```

```python
user = []
while(1):
    l = uio.readline()
    if not l:
        break
    if(direname.match(l)):
        for i in range(100+20):
            user.append(uio.readline())
```

```python
time = []
while(1):
    l = tio.readline()
    if not l:
        break
    if(direname.match(l)):
        for i in range(100+20):
            time.append(tio.readline())
            tio.readline()
            tio.readline()
```

```python
import numpy as np
```

```python
timei = np.array([float(x[4:].strip()) for x in time])
```

```python
useri = np.array([float(re.search('运行时间为 ([0-9.]+) us', x).groups()[0]) for x in user])
```

```python
timei = timei.reshape(6, -1)
useri = useri.reshape(6, -1)
```

```python
from matplotlib import pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
plt.figure(figsize=(8, 8))
for i in range(6):
    plt.plot(X, timei[i], label=keym[i])
plt.title("程序外时间测试（系统time)命令")
plt.xlabel("规模/个")
plt.ylabel("用时/s")
plt.legend()
plt.savefig("D:/Github/Parallel_Program/doc/time.jpg")
```

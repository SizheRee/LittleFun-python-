```python
"""
  @Author: Xiaosen Wang
  @Date: 2019-07-11 10:48:10
  @Last Modified by:   Xiaosen Wang
  @Last Modified time: 2019-07-11 10:48:10
  @Function: Generate the 'T' and 'L' dataset
"""
import numpy as np
import operator
import argparse
import os
from matplotlib import pyplot as plt
```

```python
"""
Generate the letter 'T' in a size * size grid
@parameter size: the size of the grid
@parameter width: the width of the letter
@return the generated letter
"""
def drawT(size, width=1):
    if (size - width) <= 0 or (size - width) % 2 != 0:
        print('Illeage parameters!')
        return
    letter = np.zeros((size, size))

    for i in range(size):
        for j in range(size):
            if i < width or (j >= (size - width) // 2 and j < (size + width) // 2):
                letter[i, j] = 1
    
    return letter
```

```python
"""
Generate the letter 'L' in a size * size grid
@parameter size: the size of the grid
@parameter width: the width of the letter
@return the generated letter
"""
def drawL(size, width=1):
    if width >= size:
        print('Illeage parameters!')
        return
    letter = np.zeros((size, size))

    for i in range(size):
        for j in range(size):
            if i >= (size - width) or (j < width):
                letter[i, j] = 1
    
    return letter
```

```python
"""
Shift the letter right by x cell and low by y cell
@parameter letter: the letter which is represented in a grid
@parameter x: the distance to be shifted right, must be in (0, 1)
@parameter y: the distance to be shifted low, must be in (0, 1)
@return the shifted letter
"""
def shift(letter, x=0, y=0):
    if x == 0 and y == 0:
        return letter
    if x >= 1 or y >= 1 or x < 0 or y < 0:
        print('Illeage parameters! (%s, %s)'%(x, y))
        return
    
    # shift the letter right by x cell
    if x > 0:
        new_letter = np.zeros(tuple(map(operator.add, letter.shape, (0, 1))))
        for i in range(new_letter.shape[0]):
            for j in range(new_letter.shape[1]):
                # the leftmost cells
                if j == 0 and letter[i, j] > 0:
                    new_letter[i, j] = letter[i, j] - x * letter[i, j]
                # the rightmost cells
                elif j == new_letter.shape[1] - 1:
                    new_letter[i, j] = letter[i, j - 1] * x
                elif j > 0:
                    new_letter[i, j] = letter[i, j] * (1 - x) + letter[i, j - 1] * x
        letter = new_letter
    
    # shift the letter low by y cell
    if y > 0:
        new_letter = np.zeros(tuple(map(operator.add, letter.shape, (1, 0))))
        for i in range(new_letter.shape[0]):
            for j in range(new_letter.shape[1]):
                # the topmost cells
                if i == 0 and letter[i, j] > 0:
                    new_letter[i, j] = letter[i, j] - y * letter[i, j]
                # the lowmost cells
                elif i == new_letter.shape[0] - 1:
                    new_letter[i, j] = letter[i - 1, j] * y
                elif i > 0:
                    new_letter[i, j] = letter[i, j] * (1 - y) + letter[i - 1, j] * y
        letter = new_letter
    
    return letter
```

```python
"""
draw the letter right in a size * size image with position (x,y)
@parameter x: the x-coordinate of the top left corner of the letter
@parameter y: the y-coordinate of the top left corner of the letter
@parameter letter: the letter which is represented in a grid
@parameter size: the size of the image
@return the shifted letter
"""
def draw(x, y, letter, size=28):
    # get the shifted letter
    letter = shift(letter, y%1, x%1)
    img = np.zeros((size, size))
    a = x
    b = y
    x = int(x // 1)
    y = int(y // 1)
    if x < 0 or y < 0 or x + letter.shape[0] > size or y + letter.shape[1] > size:
        print('x = {}, y = {}, shape = {}'.format(a, b, letter.shape))
        print('Illeage parameters!')
        return
    img[x : x + letter.shape[0], y : y + letter.shape[1]] = letter
    return img.reshape((size, size, 1))

"""
randomly generate the x-coordinate and y-coordinate in the image
@parameter img_size: the size of the image
@parameter letter_size: the size of gird which the letter is in
@return the x-coordinate and y-coordinate
"""
def sample_position(img_size, letter_size):
    return round(np.random.uniform(0, img_size - letter_size), 2), round(np.random.uniform(0, img_size - letter_size), 2)

"""
generate the dataset
@parameter train_size: the size of train data
@parameter testn_size: the size of test data
@parameter letter_size: the size of gird which the letter is in
@parameter width: the width of the letter
@parameter img_size: the size of the image
@return the train set and test set
"""
def generate_data(train_size, test_size, letter_size, width, img_size):
    letter_t = drawT(letter_size, width)
    letter_l = drawL(letter_size, width)

    x_train = np.zeros((train_size * 2, img_size, img_size, 1), np.float32)
    x_test = np.zeros((test_size * 2, img_size, img_size, 1), np.float32)

    y_train = np.zeros((train_size * 2, 2), np.int32)
    y_test = np.zeros((test_size * 2, 2), np.int32)

    y_train[:train_size, 0] = 1
    y_train[train_size:, 1] = 1
    y_test[:test_size, 0] = 1
    y_test[test_size:, 1] = 1

    for i in range(train_size):
        x, y = sample_position(img_size, letter_size)
        x_train[i] = draw(x, y, letter_t, img_size)
        x, y = sample_position(img_size, letter_size)
        x_train[i + train_size] = draw(x, y, letter_l, img_size)
    
    for i in range(test_size):
        x, y = sample_position(img_size, letter_size)
        x_test[i] = draw(x, y, letter_t, img_size)
        x, y = sample_position(img_size, letter_size)
        x_test[i + test_size] = draw(x, y, letter_l, img_size)
    
    return (x_train, y_train), (x_test, y_test)

def main():
    parser = argparse.ArgumentParser(description='Generate the dataset of letter T and L')
    parser.add_argument('--train_size', type=int, default=6000, help='the number of images of each class in train set')
    parser.add_argument('--test_size', type=int, default=1000, help='the number of images of each class in test set')
    parser.add_argument('--letter_size', type=int, default=10, help='the size of gird which the letter is in')
    parser.add_argument('--width', type=int, default=2, help='the width of the letter')
    parser.add_argument('--img_size', type=int, default=28, help='the size of the image')
    parser.add_argument('--path', type=str, default='data', help='the path to save the generated images')
    
    args = parser.parse_args()

    (x_train, y_train), (x_test, y_test) = generate_data(args.train_size, args.test_size, args.letter_size, args.width, args.img_size)

    np.savez(os.path.join(args.path, 'Letter_{}_{}'.format(args.train_size, args.test_size)), x_train=x_train, y_train=y_train, x_test=x_test, y_test=y_test)
```

```python
def d(xt, yt, xl, yl):
# xt=10.3
# yt=10.5
# xl=11.1
# yl=9.7
#     print("DEBUG", xt, yt, xl, yl)
    imgL = np.zeros((28, 28))
    imgT = np.zeros((28, 28))
    xtfc, xtdc = np.modf(xt)
    ytfc, ytdc = np.modf(yt)
    xlfc, xldc = np.modf(xl)
    ylfc, yldc = np.modf(yl)
    L_shift = shift(L, ylfc, xlfc)
    T_shift = shift(T, ytfc, xtfc)
    xlbase, ylbase = np.where(L_shift!=0)
    xtbase, ytbase = np.where(T_shift!=0)
    imgL[xlbase+np.int(xldc), ylbase+np.int(yldc)] = L_shift[xlbase, ylbase]
    imgT[xtbase+np.int(xtdc), ytbase+np.int(ytdc)] = T_shift[xtbase, ytbase]
    # imgT[xtbase + np.int(xtdc), ytbase + np.int(ytdc)] = 1
    # imgT_shift = shift(imgT, xtfc, ytfc)
    # imgT_shift_full[imgL_shift.nonzero()] = 1

#     plt.subplot(1, 2, 1)
#     plt.imshow(imgL)
#     plt.subplot(1, 2, 2)
#     plt.imshow(imgT)
    return np.sum((imgL - imgT)**2)
```

```python
def f(xt, yt, xl, yl):
    if(d(xt, yt, xl, yl) < dmax):
        return 0
    else:
        return (xt-xl)**2 + (yt-yl)**2
```

```python
def loss(xt, yt, xl, yl):
    return d(xt, yt, xl, yl) + f(xt, yt, xl, yl)
```
